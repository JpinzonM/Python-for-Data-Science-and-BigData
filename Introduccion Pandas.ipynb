Using Groupby = AGrupar por categorias. 

```python
#El retraso maximo, que ha tenido cada dia de la semana en nuestra data.
df.groupby(by= "DayOfWeek")["ArrDelay"].max()
```

```python
#media de retraso para cada dia de retraso
df.groupby(by= "DayOfWeek")["ArrDelay"].mean()
```

```python
#el retraso minimo que ha tenido cada dia de la semana.
df.groupby(by= "DayOfWeek")["ArrDelay"].min()
```

```python
df.groupby(by= "DayOfWeek")["ArrDelay"].count()
```

```python
df.groupby(by= "DayOfWeek")["ArrDelay"].describe()
```

```python
df.groupby(["DayOfWeek"])[["ArrDelay","DepDelay"]].mean()
```

```python
df.groupby("DayOfWeek")["ArrDelay"].max() - df.groupby(by= "DayOfWeek")["ArrDelay"].min()
```

```python
dfAUSHOU = df[df.Origin.isin(["AUS","HOU"])]
```

```python
#frente a los dias de la semana los vuelos que salen de las ciudades, y ver 
#o comparar cual sale mas retrasado.
dfAUSHOU.groupby(by=["DayOfWeek", "Origin"])["ArrDelay"].mean()
```

```python
dfAUSHOU.groupby(by=["Origin","DayOfWeek"])["ArrDelay"].mean()
```
