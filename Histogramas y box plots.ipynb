```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
import seaborn as sns
```

```python
#Load Data
df = pd.read_csv('C:/Users/jorge/Desktop/Linkedin DATASCIENCE/CURSO/data/2008.csv', nrows=100000)
```

```python
#Borrar datos faltantes.
#Crear subset con los datos que nos interesan
df.dropna(inplace=True, subset = ["ArrDelay", "DepDelay", "Distance"])
```

```python
#sns.distplot(df["Distance"])
sns.histplot(df["Distance"],  kde=True, stat="density", linewidth=0)
```

```python
#plot sin aproximacion de la densidad encima
sns.distplot(df["Distance"], kde = False)
```

```python
sns.histplot(df["Distance"])
```

```python
sns.distplot(df["Distance"], kde = False)
```

```python
# numero de bins de nuestro histograma, ejm: 100 cortes.
sns.distplot(df["Distance"], kde = False, bins=100)
```

```python
#comparar dos distribuciones
sns.kdeplot(df["ArrDelay"])
sns.kdeplot(df["DepDelay"])
plt.xlim([-300, 300])
```

```python
#Creacion de diagrama de caja más basico.
df2 = df[df["Origin"].isin(["ATL","HOU","IND"])].sample(frac=1).head(500)
```

```python
sns.boxplot(x="DepDelay", y = "Origin", data=df2)
```

```python
#agregando limites.
sns.boxplot(x="DepDelay", y = "Origin", data=df2)
plt.xlim([-20,150])
#la linea de la mitad del box, delimita la mediana, la mediana es el estadistico que separa  
#el 50% inferior de los casos, del 50% superior de los caso. 
#la primera linea del box, es el primer cuartil, que separa el 25% inferior de lso casos del 75% superior de los casos
#la linea del final del box, es el 3er cuartil, que separa el 75% del 25% de los casos, 
#y los puntos que parecen rombos, son los outliers que detecta nuestro boxplot, 
#basandose en el siguiente criterio de decision, todo aquel caso que super 1.5 veces el rango intercuartilico.
```
