```python
# Modificar graficos: 
#load Data 1.M
np.random.seed(0)
df = pd.read_csv('C:/Users/jorge/Desktop/Linkedin DATASCIENCE/CURSO/data/2008.csv', nrows=1000000)
df = df.sample(frac=1).head(100)
```

```python
#load Data
df = pd.read_csv('C:/Users/jorge/Desktop/Linkedin DATASCIENCE/CURSO/data/2008.csv', nrows=100000)
data = np.unique(df.DayOfWeek, return_counts=True)
labs = ["Lun", "Mar", "Mie", "Jue", "Vie", "Sab", "Dom"]
data
```

```python
plt.pie(x = data[1],
        labels= data[0],
        radius=1.5,
        colors=["Red", "Green", "Orange", "Blue", "Gray", "Pink", "Black"],
        startangle=90)
plt.show()
```

```python
#Modificar etiquetas de los triangulos
plt.pie(x = data[1],
        labels= labs,
        radius=1.5,
        colors=["Red", "Green", "Orange", "Blue", "Gray", "Pink", "Black"],
        startangle=90)
plt.show()
```

```python
#cambiar secuencia de colores por una predefinida
#del paquete seaborn.
plt.pie(x = data[1],
        labels= labs,
        radius=1.5,
        colors=sns.color_palette("hls",7),
        startangle=90)
plt.show()
```

```python
#extrae los diagramas de sectores
plt.pie(x = data[1],
        labels= labs,
        radius=1.5,
        colors=sns.color_palette("hls",7),
        explode = (0,0,.2,0,0,0,0.1),
        startangle=90)
plt.show()
```

```python
#poner los porcentajes de cada sector. 
plt.pie(x = data[1],
        labels= labs,
        radius=1.5,
        colors=sns.color_palette("hls",7),
        explode = (0,0,.2,0,0,0,0.1),
        startangle=90,
       autopct = "%1.1f%%")
plt.show()
```

```python
#label distance
plt.pie(x = data[1],
        labels= labs,
        radius=1.5,
        colors=sns.color_palette("hls",7),
        explode = (0,0,.2,0,0,0,0.1),
        startangle=90,
       autopct = "%1.1f%%",
       labeldistance=1.5)
plt.show()
```

```python
#a√±adir una legenda
plt.pie(x = data[1],
        #labels= labs,
        radius=.7,
        colors=sns.color_palette("hls",7),
        explode = (0,0,.2,0,0,0,0.1),
        startangle=90,
       autopct = "%1.1f%%")
plt.legend(loc="upper right", labels = labs)
plt.show()
```
