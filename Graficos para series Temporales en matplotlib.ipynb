```python
#Load Libraries
import pandas as pd 
import seaborn as sns 
import numpy as np
import datetime
import time
```

```python
#Load Data
df = pd.read_csv('C:/Users/jorge/Desktop/Linkedin DATASCIENCE/CURSO/data/2008.csv')
df2 = df[df["Origin"].isin(["ATL","HOU","IND"])]
df = df.head(500000)
```

```python
#preprocesamiento de datos
times = []
for i in np.arange(len(df)):
    times.append(datetime.datetime(year=2008, month = df.loc[i, "Month"], day = df.loc[i, "DayofMonth"]))
```

```python
times[50000]
```

```python
df["Time"]=times
```

```python
data = df.groupby((["Time"]), as_index=False)[["DepDelay", "ArrDelay"]].mean()
data.head()
```

```python
sns.lineplot(x = data["Time"], y = data["DepDelay"])
```

```python
data = df.groupby((["Time"]))[["DepDelay", "ArrDelay"]].mean()
data.head()
```

```python
#estructura muy practica que permite agregar mas columnas y de una muestra la legenda.
sns.lineplot(data = data)
```

```python
times = []
for i in df2.index: 
    times.append(datetime.datetime(year= 2008, month=df2.loc[i,"Month"],day=df2.loc[i,"DayofMonth"]))
    
df2["Time"] = times
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x="Time", y="ArrDelay", hue="Origin", data = df2) 
```
